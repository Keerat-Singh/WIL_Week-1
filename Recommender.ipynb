```python
import numpy as np
from keras.models import Model
from keras.layers import Input, Embedding, Flatten, Dot, Dense
from keras.utils import plot_model
from sklearn.model_selection import train_test_split

# Sample data (user_id, recipe_id, rating)
# You should replace this with your actual data
data = np.array([
    [1, 1, 5],
    [1, 2, 4],
    [2, 1, 3],
    [2, 3, 5],
    [3, 2, 4],
    [3, 3, 3],
    [4, 2, 5],
    [4, 3, 2]
])

# Extract user and recipe indices
user_ids = data[:, 0]
recipe_ids = data[:, 1]

# Normalize ratings between 0 and 1
ratings = data[:, 2] / 5.0

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    data[:,:2], ratings, test_size=0.2, random_state=42
)

# Number of unique users and recipes
num_users = len(np.unique(user_ids))
num_recipes = len(np.unique(recipe_ids))

# Embedding size for users and recipes
embedding_size = 50

# Input layers
user_input = Input(shape=(1,), name='user_input')
recipe_input = Input(shape=(1,), name='recipe_input')

# Embedding layers
user_embedding = Embedding(input_dim=num_users, output_dim=embedding_size, input_length=1)(user_input)
recipe_embedding = Embedding(input_dim=num_recipes, output_dim=embedding_size, input_length=1)(recipe_input)

# Flatten embeddings
user_flat = Flatten()(user_embedding)
recipe_flat = Flatten()(recipe_embedding)

# Dot product of user and recipe embeddings
dot_product = Dot(axes=1)([user_flat, recipe_flat])

# Fully connected layer
dense_layer = Dense(1, activation='relu')(dot_product)

# Model
model = Model(inputs=[user_input, recipe_input], outputs=dense_layer)

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit([X_train[0], X_train[1]], y_train, epochs=10, batch_size=1, validation_data=([X_test[0], X_test[1]], y_test))

# Evaluate the model
loss = model.evaluate([X_test[0], X_test[1]], y_test)
print(f'Evaluation Loss: {loss}')
```
