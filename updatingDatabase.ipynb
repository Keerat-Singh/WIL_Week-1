```python
import numpy as np

# Set the parameters
average_rating = 4
std_deviation = 1.5
max_rating = 5

# Set the number of ratings you want to generate
num_ratings = 6163  # You can change this number as needed

# Generate the rating column
ratings = np.random.normal(loc=average_rating, scale=std_deviation, size=num_ratings)

# Clip the ratings to ensure they are within the desired range (0 to max_rating)
ratings = np.clip(ratings, 0, max_rating)

# Round the ratings to one decimal place
ratings = np.round(ratings, decimals=1)

# Display the generated ratings
print("Generated Ratings:", ratings)
```

```python
db['Ratings'] = ratings
```
