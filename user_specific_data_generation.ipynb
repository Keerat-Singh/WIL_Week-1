```python
def newPrice(x):
    temp = 0
    match (x):
        case "Additive":
            temp = random.uniform(1,5)
        case "Bakery":
            temp = random.uniform(2,10)
        case "Beverage":
            temp = random.uniform(2,7)
        case "Beverage Alcoholic":
            temp = random.uniform(5,20)
        case "Cereal":
            temp = random.uniform(2,10)
        case "Dairy":
            temp = random.uniform(3,10)
        case "Dish":
            temp = random.uniform(8,20)
        case "Essential Oil":
            temp = random.uniform(4,12)
        case "Fish":
            temp = random.uniform(3,15)
        case "Flower":
            temp = random.uniform(2,10)
        case "Fruit":
            temp = random.uniform(2,6)
        case "Fungus":
            temp = random.uniform(1,4)
        case "Herb":
            temp = random.uniform(3,10)
        case "Legume":
            temp = random.uniform(1,5)
        case "Maize":
            temp = random.uniform(1,5)
        case "Meat":
            temp = random.uniform(10,20)
        case "Nuts & Seed":
            temp = random.uniform(8,15)
        case "Plant":
            temp = random.uniform(1,10)
        case "Seafood":
            temp = random.uniform(5,15)
        case "Spice":
            temp = random.uniform(1,5)
        case "Vegetable":
            temp = random.uniform(1,12)

    return round(temp, 1)
```

```python
ingredients_price_stock["New Price"] = ingredients_price_stock['Category'].apply(lambda x: newPrice(x))
```

# Popularity Score

```python
finalised_dataset.head()
```

```python
def popularity_score():
    return round(random.uniform(1, 10), 1)
```

```python
def cost():
    return round(random.uniform(7, 25), 1)
```

```python
def update_rating():
    return round(random.uniform(1, 5), 1)
```

```python
temp = finalised_dataset.copy()
```

```python
temp['Popularity Score'] = temp['Title'].apply(lambda x: popularity_score())
```

```python
temp['Rating'] = temp['Title'].apply(lambda x: update_rating())
```

```python
temp['Ingredient Cost (CAD)'] = temp['Title'].apply(lambda x: cost())
```

```python
temp.drop(columns= ['Ingredients', 'Cuisine', 'difficulty_level', 'preparation_time','rating'], inplace= True)
```
